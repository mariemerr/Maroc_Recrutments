```python
import pandas as pd
import os
```

```python
# 📂 Paths
base_path = r"C:\Users\meriem\Desktop\Recrutment\Data Clean"

raw_path = os.path.join(base_path, "all_data_raw.csv")
clean_path = os.path.join(base_path, "all_data_clean.csv")
filtered_path = os.path.join(base_path, "all_data_clean_2021_2025.csv")
```

```python
df = pd.read_csv(raw_path, encoding="utf-8-sig")
```

```python
# 2️⃣ Nettoyage des dates
date_cols = [
    "تاريخ النشر",
    "آخر أجل لإيداع الترشيحات",
    "تاريخ الإمتحان"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce").dt.date

print("📅 Dates nettoyées (sans heure)")

# Test : affichage des 5 premières lignes des colonnes de dates
print("\n👀 Aperçu des dates nettoyées :")
print(df[date_cols].head())
```

```python
# 2️⃣ Remplir les valeurs manquantes
df["Corps"] = df["Corps"].fillna("Non spécifié")
df["Grade"] = df["Grade"].fillna("Non spécifié")
df['Nom du poste'] = df['Nom du poste'].fillna('Non spécifié')
df['عدد المناصب'] = df['عدد المناصب'].fillna(0).astype(int)

print("🧩 Valeurs manquantes corrigées")

# 📝 Test : afficher les 5 premières lignes des colonnes corrigées
print("\n👀 Aperçu des valeurs après remplissage :")
print(df[["Corps", "Grade"]].head(10))
```

```python
# 3️⃣ Conversion عدد المناصب إلى int
df["عدد المناصب"] = (
    pd.to_numeric(df["عدد المناصب"], errors="coerce")
    .fillna(0)
    .astype(int)
)

print("🔢 عدد المناصب حُوّل إلى أعداد صحيحة")
```

```python
# 4️⃣ Nettoyage des colonnes texte
text_cols = [
    "Administration",
    "Nom du poste",
    "Corps",
    "Grade",
    "Type_Source"
]

for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

print(" Texte nettoyé (espaces supprimés)")

# Test : affichage des 
print("\n Aperçu du texte nettoyé :")
print(df[text_cols].head())
```

```python
# 5  Extraire l'année à partir de تاريخ النشر
df['Année'] = pd.to_datetime(df['تاريخ النشر'], errors='coerce').dt.year

# 📝 Test : afficher les 5 premières lignes avec l'année
print("\n📆 Aperçu avec la colonne Année :")
print(df[['تاريخ النشر', 'Année']].head())
```

```python
# 7️⃣ Sauvegarde finale dans Data Clean
df.to_csv(clean_path, index=False, encoding="utf-8-sig")

print("✅ Nettoyage final terminé")
print("📊 Dimensions finales :", df.shape)
```

```python
# 8️⃣ Filtrage années 2021–2025
annees_valides = [2021, 2022, 2023, 2024, 2025]

df = df[df["Année"].isin(annees_valides)]
print("\n📆 السنوات بعد التصفية :")
print(sorted(df["Année"].dropna().unique()))

print("\n📊 عدد السطور بعد التصفية :", df.shape[0])
print("\n📈 عدد الإعلانات حسب السنة :")
print(df["Année"].value_counts().sort_index())

df.to_csv(filtered_path, index=False, encoding="utf-8-sig")
print("✅ Dataset filtré (2021–2025) sauvegardé avec succès")
```

```python

# 9️⃣ Lire le dataset filtré pour vérification
df = pd.read_csv(filtered_path, encoding="utf-8-sig")

print("Dimensions :", df.shape)
print("\nColonnes :")
print(df.columns)
print("\nInfo générale :")
print(df.info())
```

```python
# 10️⃣ Supprimer les lignes où عدد المناصب = 0
nb_zero = (df["عدد المناصب"] == 0).sum()
print("🔎 Avant suppression - lignes avec عدد المناصب = 0 :", nb_zero)
print("📦 Shape avant :", df.shape)

df = df[df["عدد المناصب"] != 0].copy()

# Vérifier après suppression
print("✅ Après suppression - lignes avec عدد المناصب = 0 :", (df["عدد المناصب"] == 0).sum())
print("📦 Shape après :", df.shape)
```

```python
# 8️⃣ Classification niveau_apres_bac
def get_niveau_final_v4(row):
    texte = " ".join([str(row.get("Grade", "")), str(row.get("Corps", "")), str(row.get("Nom du poste", ""))]).lower()
    if any(k in texte for k in ["maître de conférences","maitre de conférences","professeur de l'enseignement supérieur assistant","pesa","enseignants-chercheurs","enseignants chercheurs","doctorat"]):
        return "Doctorat (Bac+8)"
    if any(k in texte for k in ["médecin","medecin","pharmacien","chirurgien dentiste","dentiste","vétérinaire","veterinaire","interne","résidant","resident"]):
        return "Bac+7 (Doctorat)"
    if any(k in texte for k in ["ingénieur","ingenieur","architecte","master","bac+5"]):
        return "Bac+5"
    if "administrateur" in texte:
        if any(k in texte for k in ["2ème grade","2eme grade"]):
            return "Bac+5"
        else:
            return "Bac+3"
    if any(k in texte for k in ["infirmier","sage-femme","sage femme","kinésithérapeute","assistant médico-social","technicien de santé","infirmiers et techniciens de santé","isits","ispits","attaché administratif principal"]):
        return "Bac+3"
    if any(k in texte for k in ["technicien","adjoint technique","agent de maîtrise","rédacteur","redacteur","téléconseiller","dut","bts","4ème grade","4eme grade"]):
        return "Bac+2"
    if any(k in texte for k in ["gendarme","gardien","surveillant","protection civile","agent d'autorité","stagiaire de l'institut","élève agent","adjoint administratif"]):
        return "Bac"
    return "Non classifié"

df["niveau_apres_bac"] = df.apply(get_niveau_final_v4, axis=1)

# ✅ Aperçu final avant sauvegarde
colonnes_finales = ["Administration", "عدد المناصب", "Grade", "Nom du poste", "Année", "niveau_apres_bac"]
print(df[colonnes_finales].head(10))
print(df["niveau_apres_bac"].value_counts())

# --- 2. DIAGNOSTIC : Wach l'probleme ? ---
print("⚠️ Diagnostic :")
print(df["niveau_apres_bac"].value_counts())

# Hna nchoufou chnou fih "Non classifié" bach n3rfou ach m3a 9edena
print("\n🔍 Exemples de lignes 'Non classifié' :")
print(df[df["niveau_apres_bac"] == "Non classifié"][["Grade", "Corps"]].head(5))

# --- 3. ANALYSE : Somme des Postes (Le Vrai Chiffre) ---
# Hada howa l'analyse lmohimma bash t3rf ch7al men poste f kola niveau
analyse_postes = df.groupby("niveau_apres_bac")["عدد المناصب"].sum().sort_values(ascending=False)

print("\n📊 --- NOMBRE TOTAL DE POSTES PAR NIVEAU ---")
print(analyse_postes)
```

```python
# 1. Kandewzo les colonnes l-liste bach ma ytl3ch error
colonnes_pour_test = list(df.columns)

# 2. Kanjmo3ou w kan7ssbo 'count'
check_repetition = df.groupby(colonnes_pour_test).size().reset_index(name='count')

# 3. Kandewzou l-resultat l-variable 'doublons'
# Hna kanselectiw ghi les lignes li 'count' dyalha kber men 1 (ya3ni tkrarat)
doublons = check_repetition[check_repetition['count'] > 1]

# 4. Affichage
print(f"🔎 Total dyal les lignes uniques: {len(check_repetition)}")
print(f"⚠️ Nombre dyal les lignes li tkrarat (Doublons): {len(doublons)}")

if len(doublons) > 0:
    print("\n👀 Liste des lignes répétées :")
    print(doublons.sort_values(by='count', ascending=False))
else:
    print("\n✅ Conclusion: Makayn TA HAJA mtdoubla. Data ndifa !")
```
